```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
(x_train,y_train),(x_test,y_test) = keras.datasets.mnist.load_data()
```

```python
x_train.shape #60,000 data of array 28,28
```

```python
x_train[0] #pixel values
```

```python
x_test
```

```python
y_test
```

```python
import matplotlib.pyplot as plt
print(y_train[3])
plt.imshow(x_train[3])
```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
model = Sequential()
#firslty we need to flatten it...
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(10, activation = 'softmax'))
```

```python
model.summary()
```

```python
model.compile(loss ="sparse_categorical_crossentropy", optimizer ='Adam') #no need for one hot encoding,,, categorical_crossentrophy pr encode krne ki need hoti hai
```

```python
history = model.fit(x_train, y_train, epochs =15, validation_split=0.2)
```

```python
y_prob=model.predict(x_test) #each array tells the probability of the digit it can be
```

```python
y_pred = y_prob.argmax(axis = 1)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```
